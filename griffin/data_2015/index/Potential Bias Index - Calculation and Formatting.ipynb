```python
import numpy as np
import pandas as pd
import os
```

```python
m_list = []
t_list = []
i=0
for file in os.listdir('.'):
    if file.endswith('.csv'):
        df = pd.read_csv(file, sep=';')
        frame = df.copy()
        frame = frame.drop('White', axis=1)

        for index,label in enumerate(frame.Cluster):
            if 'Euclidean' in label:
                frame = frame.drop(index);


        frame['Minority Prob'] = frame.loc[:,['Black or African American', 'American Indian or Alaskan Native',
                   'Asian','Native Hawaiian or other Pacific Islander', 'Multiple Race',
                   'Other Race']].sum(axis=1)
        frame = frame.drop(['Black or African American', 'American Indian or Alaskan Native',
                   'Asian','Native Hawaiian or other Pacific Islander', 'Multiple Race',
                   'Other Race', 'TN','TP'], axis =1)

        frame['Misclassifications'] = frame.loc[:,['FP', 'FN']].sum(axis=1)
        frame = frame.drop(['FP','FN'], axis =1)
        frame['Potential Bias Index'] = frame.loc[:,'Minority Prob'] * frame.loc[:,'Misclassifications']


        out = frame.loc[:,'Potential Bias Index'].mean()
        if(i < 12):
            m_list.append(out)
        else:
            t_list.append(out)
        
        print(f'{i%12} File:{file} - {out}')
        i +=1
        
        
# This is gross but the easiest way to rearrange the list
m_list = [m_list[4],m_list[3],m_list[7],m_list[0],m_list[8],
         m_list[6],m_list[5],m_list[1],m_list[11],m_list[10],
         m_list[9],m_list[2]]

t_list = [t_list[4],t_list[3],t_list[7],t_list[0],t_list[8],
         t_list[6],t_list[5],t_list[1],t_list[11],t_list[10],
         t_list[9],t_list[2]]


writer = open('pbIndex.js','w')
writer.write('var motor =['+ ','.join(map(str,m_list)) + ']' +  '\n')
writer.write('var theft =['+ ','.join(map(str,t_list)) + ']' +  '\n')
writer.close()
frame
```
